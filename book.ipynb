### set the min `number of ratings` by an user is 200

```python
ratings['user_id'].value_counts().shape
```

```python
ratings['user_id'].value_counts()[ratings['user_id'].value_counts()>200]
```

```python
x = ratings['user_id'].value_counts()
y = x[x>200]
y
```

```python
y.index
```

```python
ratings = ratings[ratings['user_id'].isin(y.index)]
ratings
```

```python
ratings_with_books= ratings.merge(books,on='ISBN')
ratings_with_books
```

```python
ratings_with_books['title'].nunique()
```

```python
ratings_with_books['ISBN'].nunique()
```

```python
no_of_ratings = ratings_with_books.groupby('title')['rating'].count().reset_index()
```

```python
no_of_ratings.rename(columns={'rating':'num_ratings'},inplace=True)
no_of_ratings
```

```python
final_ratings = ratings_with_books.merge(no_of_ratings,on='title')
final_ratings
```

### `num_ratings` is being set to >=50

```python
final_ratings = final_ratings[final_ratings['num_ratings']>=50]
final_ratings.shape
```

```python
final_ratings.drop_duplicates(['user_id','title'],inplace=True)
```

```python
final_ratings
```

```python
final_ratings['title'] = final_ratings['title'].str.replace("\\","")
```

```python
#book_pivot = final_ratings.pivot(columns='title',index='user_id',values='rating')
```

```python
book_pivot = final_ratings.pivot(columns='user_id',index='title',values='rating')
```

```python
book_pivot
```

```python
book_pivot.fillna(0,inplace=True)
```

```python
from scipy.sparse import csr_matrix
book_sparse = csr_matrix(book_pivot)  # reverse is `book_pivot = book_sparse.toarray()`
```

```python
type(book_sparse)
```

```python
from sklearn.neighbors import NearestNeighbors

model = NearestNeighbors(algorithm = 'brute')
```

```python
model.fit(book_sparse)
```

```python
def recommend(book_name):
    book_index = np.where([book_pivot.index==book_name])[1][0]
    distances, suggestions = model.kneighbors(book_pivot.iloc[book_index,:].values.reshape(1,-1),n_neighbors=6)
    
    for i in range(len(suggestions[0])):
        index=suggestions[0][i]
        book = book_pivot.index[index]
        print(book)
 
```

```python
def recommend(book_name):
    book_index = np.where([book_pivot.index==book_name])[1][0]
    distances, suggestions = model.kneighbors(book_pivot.iloc[book_index,:].values.reshape(1,-1),n_neighbors=6)
    
    books=[]
    for i in range(len(suggestions[0])):
        index=suggestions[0][i]
        book = book_pivot.index[index]
        books.append(book)
    return books
 
```
